```python
from simple_data_analysis.config import RAW_DATA_DIR
import scipy.stats as st
import numpy as np
import pandas as pd

VERSION = '1'
dir = RAW_DATA_DIR / VERSION
```

# 8.1
<h3 style="direction: ltr;">میانگین طول آهنگ ها در ژانر های مختلف یکسان؟ با فاصله اطمینان 95 درصدی</h3>
<p style="text-alig: left: direction: ltr;">خیر</p>

```python
track = pd.read_csv(dir / "Track.csv")
genre = pd.read_csv(dir / "Genre.csv")

track = track.set_index(keys=['TrackId'])
genre = genre.set_index(keys=['GenreId'])

track = track.join(genre, how='left', on='GenreId', lsuffix='_x', rsuffix='_y')
track = track.rename(columns={'Name_x': 'Name', 'Name_y': 'Genre'})
track["Second"] = track['Milliseconds'] / 1000

df = track.groupby(by='Genre')

result = {
    "Genre": [],
    "Mean": [],
    "Interval_I": [],
    "Interval_J": []
}

for GenreName, data in df:
    if len(data['Second']) - 1 == 0:
        continue

    i, j = st.t.interval(0.95, len(data['Second']) - 1, loc=np.mean(data['Second']), scale=st.sem(data['Second']))
    result['Genre'].append(GenreName)
    result['Mean'].append(data[(data['Second'] >= i) & (data['Second'] <= j)]['Second'].mean())
    result['Interval_I'].append(i)
    result['Interval_J'].append(j)


df = pd.DataFrame(result)
df
```
